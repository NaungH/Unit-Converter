```python
# Miles to Kilometers
 
miles = float(input("Enter the Miles = "))
 
kilometers = miles * 1.6093435
 
print("%.2f Miles equals %.2f Kilometers " %(miles, kilometers))
```

```python
# feet to meters

feet=float(input("Enter the feet: "))

meter = feet * 0.305

print(feet,"feet equal to ",meter,"meters")
```

```python
#'F to 'C
F = float(input("Enter the 'F "))

C = F * 33.8

print (F, "'F equal to ", C, "'C")
```
